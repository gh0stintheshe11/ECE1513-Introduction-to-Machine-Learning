```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Given dataset
X = np.array([[1, 1], [-1, -1], [1, 0], [0, 1]])  # Feature points
y = np.array([1, -1, -1, -1])  # Class labels

# Train SVM classifier
svc = SVC(kernel='linear', C=1e6)  # Hard margin SVM
svc.fit(X, y)

# Extract the weight vector and bias
w = svc.coef_[0]
b = svc.intercept_[0]

# Define decision boundary and margin lines
x_vals = np.linspace(-1.5, 1.5, 100)
y_decision = -(w[0] * x_vals + b) / w[1]
y_margin_pos = -(w[0] * x_vals + b - 1) / w[1]
y_margin_neg = -(w[0] * x_vals + b + 1) / w[1]

# Plot dataset
plt.figure(figsize=(6, 6))
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label="Class +1", edgecolors='k', s=100)
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='blue', label="Class -1", edgecolors='k', s=100, marker='x')

# Plot decision boundary and margins
plt.plot(x_vals, y_decision, 'k-', label="Decision Boundary")
plt.plot(x_vals, y_margin_pos, 'r--', label="Margin +1")
plt.plot(x_vals, y_margin_neg, 'b--', label="Margin -1")

# Labels and legend
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.title("SVC Dataset with Maximum-Margin Line")
plt.legend()
plt.grid()
plt.show()
```
